```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
train_df = pd.read_excel('/content/drive/MyDrive/Flight Prediction/Data_Train.xlsx')
train_df.head()
```

```python
test_df = pd.read_excel('/content/drive/MyDrive/Flight Prediction/Test_set.xlsx')
test_df.head()
```

```python
# combine two dataset
final_df = train_df.append(test_df)
final_df.head()
```

```python
final_df.info()
```

```python
final_df['Date_of_Journey'].str.split('/')
```

```python
# feature engineering process
final_df['Date'] = final_df['Date_of_Journey'].str.split('/').str[0]
final_df['Month'] = final_df['Date_of_Journey'].str.split('/').str[1]
final_df['Year'] = final_df['Date_of_Journey'].str.split('/').str[2]
```

```python
final_df.head(1)
```

```python
final_df['Date'] = final_df['Date'].astype(int)
final_df['Month'] = final_df['Month'].astype(int)
final_df['Year'] = final_df['Year'].astype(int)
```

```python
final_df.drop('Date_of_Journey', axis=1, inplace=True)
```

```python
final_df['Arrival_Time'] = final_df['Arrival_Time'].apply(lambda x : x.split(' ')[0])
```

```python
final_df['Arrival_hour'] = final_df['Arrival_Time'].str.split(':').str[0]
final_df['Arrival_min'] = final_df['Arrival_Time'].str.split(':').str[1]
final_df['Arrival_hour'] = final_df['Arrival_hour'].astype(int)
final_df['Arrival_min'] = final_df['Arrival_min'].astype(int)
final_df.drop('Arrival_Time', axis=1, inplace=True)
final_df.head(2)
```

```python
final_df['Dep_hour'] = final_df['Dep_Time'].str.split(':').str[0]
final_df['Dep_min'] = final_df['Dep_Time'].str.split(':').str[1]
final_df['Dep_hour'] = final_df['Dep_hour'].astype(int)
final_df['Dep_min'] = final_df['Dep_min'].astype(int)
final_df.drop('Dep_Time', axis=1, inplace=True)
final_df.head(2)
```

```python
final_df.info()
```

```python
dic_1 = {
    'non-stop':0,
    '2 stops':1,
    '1 stop':2,
    '3 stops':3,
    'nan':np.nan,
    '4 stops':4
}
final_df['Total_Stops'] = final_df['Total_Stops'].map(dic_1)
```

```python
final_df.drop('Route', axis=1, inplace=True)
```

```python
final_df.head()
```

```python
final_df['Additional_Info'].unique()
```

```python
#convert duration into minit 
final_df['duration_hour'] = final_df['Duration'].str.split(' ').str[0].str.split('h').str[0]
final_df['duration_min'] = final_df['Duration'].str.split(' ').str[1].str.split('m').str[0]
```

```python
final_df.drop(final_df[final_df['duration_hour'] == '5m'].index.tolist(), axis=0, inplace=True)
```

```python
final_df['duration_hour'] = final_df['duration_hour'].astype(int)
final_df['duration_hour'] = [i*60 for i in final_df['duration_hour']]
```

```python
final_df['Airline'].unique()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
final_df['Airline'] = le.fit_transform(final_df['Airline'])
final_df['Source'] = le.fit_transform(final_df['Source'])
final_df['Destination'] = le.fit_transform(final_df['Destination'])
final_df['Additional_Info'] = le.fit_transform(final_df['Additional_Info'])
```

```python
final_df.head()
```

```python
final_df.drop('Duration', axis=1, inplace=True)
```
