```python
countries = [
             'USA',
             'MOROCCO',
             'FRANCE',
             'RUSSIA',
             'ALGERIA'
]
print('Countries : ',countries)
# indexing / slicing
print(
    countries[0], #index
    countries[-1], # negative index
    countries[2:4], # slicing (list subset)
    countries[2:], # slicing (list subset)
    countries[:4], # slicing (list subset)
)
# setting
countries.append('CANADA') #adding
countries.insert(0, 'SPAIN') #adding in a location
countries_2 = ['UK', 'GERMANY', 'AUSTRIA'] # join lists
nested_list = [countries, countries_2] #list inside list

print(
    countries+countries_2,'\n',
    nested_list
)
# remove
countries.remove('USA')
countries.pop(-1) # take index, pop element and return it
del countries[0]

# sorting
numbers = [i for i in range(10)] # create a sorted list
numbers.sort()
print(numbers)
numbers.sort(reverse=True)
print(numbers)

#update
numbers[0] = 10

# copy
new_list = countries[:] #from 0 to the end
new_list_2 = countries.copy() #create copy
```
